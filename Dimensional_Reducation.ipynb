```python
from tensorflow.keras.datasets import mnist
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import umap
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test,y_test) = mnist.load_data()
print(x_train.shape, y_train.shape,x_test.shape,y_test.shape)
```

```python
x_train = x_train.reshape(len(x_train),-1)
x_train.shape
```

```python
#PCA
pca = PCA(n_components=2).fit_transform(x_train)
```

```python
plt.figure(figsize = (10,8))
c = plt.scatter(pca[:,0],pca[:,1], c=y_train  )
plt.legend(handles=c.legend_elements()[0],labels=list(range(10)))
plt.show
```

```python
#LDA
lda = LDA(n_components=2).fit_transform(x_train,y_train)
```

```python
plt.figure(figsize=(10,8))
sc = plt.scatter(lda[:,0],lda[:,1],c=y_train)
plt.legend(handles=sc.legend_elements()[0],labels=list(range(10)))
plt.show()
```

```python
#TSNE
tsne = TSNE().fit_transform(x_train)
```

```python
plt.figure(figsize=(10,10))
sc = plt.scatter(tsne[:,0],tsne[:,1],c=y_train)
plt.lengent(handles=sc.langend_elemnents()[0],labels=list(range(10)))
plt.show
```

```python
#UMAP
umap = UMAP(n_neighbourd=10,min_dist=0.1,metric='correlation')
plt.figure(figsize=(10,10))

sc = plt.scatter(umap[:,0],umap[:,1],c=y_train)
plt.legend(handles=sc.lagend_elementd()[0],labels=list(range(10))
plt.show()
```
